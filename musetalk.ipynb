```python
import os
```

```python
result_dir= "/kaggle/working/results/"
audio_path = "/kaggle/input/m/polinakuleshova/musetalk/pytorch/default/1/MuseTalk/data/audio/yongen.wav"
video_path = "/kaggle/input/m/polinakuleshova/musetalk/pytorch/default/1/MuseTalk/data/video/yongen.mp4"
input_basename = os.path.basename(video_path).split('.')[0]
audio_basename  = os.path.basename(audio_path).split('.')[0]
output_basename = f"{input_basename}_{input_basename}"
result_img_save_path = os.path.join(result_dir, output_basename)
output_vid_name = os.path.join(result_dir, output_basename+".mp4")
fps = 25
```

```python
!python -m scripts.inference --inference_config configs/inference/test.yaml --result_dir=/kaggle/working/results/
```

```python
result_img_save_path
```

```python
import os
import shutil
cmd_img2video = f"ffmpeg -y -v warning -r {fps} -f image2 -i {result_img_save_path}/%08d.png -vcodec libx264 -vf format=rgb24,scale=out_color_matrix=bt709,format=yuv420p -crf 18 /kaggle/working/results/temp.mp4"
print(cmd_img2video)
os.system(cmd_img2video)
cmd_combine_audio = f"ffmpeg -y -v warning -i {audio_path} -i /kaggle/working/results/temp.mp4 {output_vid_name}"
print(cmd_combine_audio)
os.system(cmd_combine_audio)
os.remove("/kaggle/working/results/temp.mp4")
shutil.rmtree(result_img_save_path)
print(f"result is save to {output_vid_name}")
```
